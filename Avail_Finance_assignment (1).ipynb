```python
#Replacing by the mode
df['REASON'].replace(np.NaN,'DebtCon',inplace=True)
```

```python
df['JOB'].value_counts()
```

```python
#Repalcing by mode
df['JOB'].replace(np.NaN,'Other',inplace=True)
```

```python
df['BAD'].value_counts()
```

```python
#Repalcing by median
df['MORTDUE'].replace(np.NaN,df['MORTDUE'].median(),inplace=True)
df['VALUE'].replace(np.NaN,df['VALUE'].median(),inplace=True)
df['YOJ'].replace(np.NaN,df['YOJ'].median(),inplace=True)
df['DEROG'].replace(np.NaN,df['DEROG'].median(),inplace=True)
df['DELINQ'].replace(np.NaN,df['DELINQ'].median(),inplace=True)
df['CLAGE'].replace(np.NaN,df['CLAGE'].median(),inplace=True)
df['NINQ'].replace(np.NaN,df['NINQ'].median(),inplace=True)
df['CLNO'].replace(np.NaN,df['CLNO'].median(),inplace=True)
```

```python
# Replace by mean
df['DEBTINC'].replace(np.NaN,df['DEBTINC'].mean(),inplace=True)
```

### Feature Engineering

```python
cf['new_1']=cf['DELINQ']/(cf['CLAGE']+1)
```

```python
cf['new_2']=(cf['DEROG']+cf['DELINQ'])/2
```

```python
cf['new_3']=cf['NINQ']/(cf['CLAGE']+1)
```

```python
cf['new_4']=np.sqrt(cf['YOJ']*cf['CLNO'])
```

```python

cf['new_5']=np.log(cf['CLNO']+1)
```

### Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
rfe = RFE(model, 4)
fit = rfe.fit(cf, label)
(print("Num Features: %d" % fit.n_features_,))
(print("Selected Features: %s" % fit.support_,))
(print("Feature Ranking: %s" % fit.ranking_,))
```

```python
#printing features selected by RFE
for i in range(len(fit.support_)):
    if fit.support_[i]:
        print(cf.columns[i])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cf, label, stratify=label, test_size=0.33,random_state=0)# 67% is for the trainingset and 33% for the validation.
```

### Feature importance using decision tree

### Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

### Modelling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

### 1. Logistic Regression


```python

clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred_lr = clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
accuracy=accuracy_score(y_test,y_pred_lr)
accuracy
```

### 2. Decision Tree Classifier

```python
# Set the parameters by cross-validation
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.33, random_state=0)

# Tree
param_grid = {"max_depth": np.linspace(10, 15, 6).astype(int),
              "min_samples_split": np.linspace(2, 5, 4).astype(int)
              }
clf_dt = GridSearchCV(DecisionTreeClassifier(), param_grid=param_grid, cv=cv)
clf_dt.fit(X_train, y_train)

y_pred_dt = clf_dt.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_dt)
accuracy=accuracy_score(y_test,y_pred_dt)
accuracy
```

### 3. Random Forest Classifier

```python
clf2=RandomForestClassifier()
clf2.fit(X_train,y_train)
y_pred_rf=clf2.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
accuracy=accuracy_score(y_test,y_pred_rf)
accuracy
```

```python
cm = confusion_matrix(y_test, y_pred_xgb)
accuracy=accuracy_score(y_test,y_pred_xgb)
accuracy
```
